```python
import matplotlib.pyplot as plt
import random


vx = [1.0, 2.0, 3.0]
vy = [1.0, 2.0, 1.0]

sx = vx[0]
sy = vy[0]

px = []
py = []

for i in range(1, 1000):
    r = random.randint(0, 2)
    sx = (sx + vx[r]) / 2
    sy = (sy + vy[r]) / 2
    px.append(sx)
    py.append(sy)
    
plt.plot(px, py, 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random


count = 4
x = list(range(1, 7))
y = [0 for i in x]

fig, ax = plt.subplots()

for _ in range(count):
    y[random.randint(0,5)] += 1
    
ax.bar(x, y)
y.sort()
stat = [f'{i + 1}: {round(y[i] / count * 100)}%' for i in range(0,6)]
print(stat)
plt.show()
```
